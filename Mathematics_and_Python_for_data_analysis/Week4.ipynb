# Гистограмма выборки

```python
%matplotlib inline
import matplotlib.pyplot as plt
import scipy.stats as sts
import numpy as np
```

```python
rv = sts.expon() 
sample = rv.rvs(1000)
```

```python
x = np.linspace(0,15,1000) #теоретическая плостность вероятности
pdf = rv.pdf(x) 
```

```python
plt.plot(x, pdf, linewidth = 2.5) #гистограма и теор. оценка плотн. вероятности
plt.hist(sample, density=True, stacked=True)
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
```

#### гистограмма выборки хорошо совпадает с теоретической оценкой плотности вероятности

## Распределение выборочных средних

```python
def find_pdf(n):  #выборочные средние по выборкам 
    means = []
    for _ in range(1000):
        sample = rv.rvs(n)
        sample_mean = np.mean(sample)
        means.append(sample_mean)
    
    # расчитаем теоритеческую плотность вероятности 
    mu = 1 
    sigma = (1./n) ** 0.5 
    norm_rv = sts.norm(loc=mu, scale=sigma)
    x = np.linspace(0,4,100)
    pdf = norm_rv.pdf(x)
    
    plt.plot(x, pdf, linewidth = 2.5)
    plt.hist(means, density=True, stacked=True)
    plt.xlabel('$\\bar{X}_n$')
    plt.ylabel('$f(\\bar{X}_n)$')
    plt.title('n = %d' % n)
    plt.show()
```

```python
for n in [3, 5, 10, 50, 200]:
    find_pdf(n)
```

### чем больше объем выборки n, тем лучше распределение выборочных средних соответствует приближающему его нормальному распределению.
